```python
import numpy as np
import pandas as pd
import pandas_profiling
import matplotlib.pyplot as plt
import seaborn as sns
from pydataset import data
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler

%matplotlib inline
```

```python
auto_df = data('mpg')
auto_df.head()
```

# Z Scaling

```python
auto_df['cty_zscore'] = (auto_df['cty'] - auto_df['cty'].mean())/auto_df['cty'].std()
auto_df.head()
```

# min-max encoding to 0,1

```python
min_max_scaler = MinMaxScaler()
scaled_array = min_max_scaler.fit_transform(auto_df['cty'].values.astype(float).reshape(-1, 1))
sa_df = pd.DataFrame(scaled_array)
sa_df = sa_df.rename(columns={0:'cty_scale'})
print(sa_df.head())

sa_df.index += 1  # index started at 1 for autoModelDF, but 0 for the hashedfeatures, correct offset
auto_df = pd.concat([auto_df,sa_df], axis=1)
auto_df.head()
```
